# Описание задачи

Представьте, что вы работаете в крупном магазине спортивных товаров. Ваша задача — с помощью данных о покупках клиентов и их социально-демографических признаках проанализировать эффективность уже проведённых ранее маркетинговых кампаний и выявить факторы, способные повысить продажи.

Вам предоставили данные о покупках клиентов за два месяца. Чтобы не возникло утечки, вся информация была зашифрована. Данные хранятся в базе данных (shop_database.db) в data.zip (папка result_data).

База данных содержит три таблицы:

- personal_data — ID клиентов, их пол, возраст, образование, страна и город проживания;
- personal_data_coeffs — данные с персональными коэффициентами клиентов, которые рассчитываются по некоторой закрытой схеме (вам потребуется коэффициент personal_coef);
- purchases — данные о покупках: ID покупателя, название товара, цвет, стоимость, гендерная принадлежность потенциальных покупателей товара, наличие скидки (поле base_sale. Значение 1 соответствует наличию скидки на момент покупки) и дата покупки.

**Обратите внимание:** страна и город проживания присваиваются человеку в соответствии с местом рождения и могут не соответствовать его реальному месту пребывания. Перед началом работы вам необходимо отфильтровать данные и оставить только тех людей, которые относятся к стране с кодовым цифровым значением 32. Все остальные неточности в данных не повлияют на качество выполнения задания.

В некоторых столбцах таблицы содержатся пропуски, информация о названии товаров неоднородна, а в данных о цветах попадаются наборы цветов, записанные через косую черту (/).

```python
def describe_df(df, name='DataFrame', show_head=True, n_head=5):
    print(f"\nАнализ датафрейма: {name}\n{'-'*40}")
    print(f"Размер: {df.shape[0]} строк × {df.shape[1]} столбцов")
    print(f"Всего пропусков: {df.isna().sum().sum()}")
    
    nulls = df.isna().sum()
    nulls = nulls[nulls > 0]
    if not nulls.empty:
        print("\nПропуски по столбцам:")
        display(nulls.to_frame('Количество пропусков'))
    
    dups = df.duplicated().sum()
    print(f"\nКоличество дубликатов: {dups}")

    print("\nТипы данных:")
    display(pd.DataFrame(df.dtypes, columns=['Тип']))

    if show_head:
        print(f"\nПервые {n_head} строк:")
        display(df.head(n_head))
```

```python
df_model = df_combined.merge(purchase_stats[['id', 'purchase_count', 'discount_rate']], on='id', how='left')
```

```python
# --- Функция для определения пола товара по названию ---
def infer_product_sex(product_name):
    if pd.isna(product_name):
        return -1  # Пропуск

    product_name = product_name.strip().lower()  # Нормализация

    # Ключевые слова по категориям
    male_keywords = ['мужск']
    female_keywords = ['женск']
    kids_keywords = ['детск', 'девоч', 'мальч']
    unisex_keywords = ['унисекс', 'универсаль']

    # Проверка по приоритету
    if any(kw in product_name for kw in male_keywords):
        return 1  # Мужской товар
    elif any(kw in product_name for kw in female_keywords):
        return 0  # Женский товар
    elif any(kw in product_name for kw in kids_keywords):
        return 3  # Детский товар
    elif any(kw in product_name for kw in unisex_keywords):
        return 4  # Унисекс
    else:
        return -1  # Неопределён
```

```python
df_purchases_cleaned['product_sex_filled'] = df_purchases_cleaned['product'].apply(infer_product_sex)
```

# Модель по восстановлению пола

```python
# --- 1. Отделяем обучающие данные, где пол известен ---

# Создаём копию, чтобы избежать изменения основного датафрейма
df_train_gender = df_model[df_model['gender'].notna()].copy()

# Целевая переменная — пол
y = df_train_gender['gender'].astype(int)

# Признаки, которые будем использовать
features = ['age', 'education', 'purchase_count', 'discount_rate']
X = df_train_gender[features]
```

```python
# --- 2. Создаём трансформер для кодирования education ---
# education — категориальный, остальные — числовые (пропускаем)

column_transformer = ColumnTransformer(
    transformers=[
        ('edu_ohe', OneHotEncoder(handle_unknown='ignore'), ['education'])
    ],
    remainder='passthrough'  # остальные признаки передаём как есть (age, purchase_count, discount_rate)
)

# --- 3. Разбиваем выборку на обучающую и тестовую ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y  # стратифицированное деление по полу
)

# --- 4. Применяем трансформер к признакам ---
X_train_transformed = column_transformer.fit_transform(X_train)
X_test_transformed = column_transformer.transform(X_test)
```

```python
# --- 5. Обучаем модель случайного леса ---
clf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'
)
clf.fit(X_train_transformed, y_train)

# --- 6. Делаем предсказания на тестовой выборке ---
y_pred = clf.predict(X_test_transformed)

# --- 7. Оцениваем метрики качества ---
print("F1-score:", f1_score(y_test, y_pred))
print("\nКлассификационный отчёт:\n", classification_report(y_test, y_pred))
print("\nМатрица ошибок:\n", confusion_matrix(y_test, y_pred))
```

## Результаты классификации

- F1-score (макро): ~0.62 — достойно для задачи восстановления пола по косвенным признакам.
- Класс 1 (женщины) классифицируется лучше, чем класс 0.
- Классы сбалансированы (поддержка: 9469 vs 12728), но precision и recall чуть лучше у класса 1.
- Матрица ошибок показывает, что модель не переобучена на один класс, и есть разумное разделение.

```python
# --- 8. Отбираем строки, где пол неизвестен ---
df_missing_gender = df_model[df_model['gender'].isna()].copy()

# Только нужные признаки
X_missing = df_missing_gender[features]

# Трансформируем с помощью уже обученного трансформера
X_missing_transformed = column_transformer.transform(X_missing)

# --- 9. Предсказываем пол ---
gender_pred = clf.predict(X_missing_transformed)

# Сохраняем предсказания в новый столбец
df_model.loc[df_model['gender'].isna(), 'gender_predicted'] = gender_pred
```

# Кластеризация

```python
# --- Подготовка датафрейма с двумя бинарными признаками образования ---
df_clusters = df_model[[
    'age', 'education', 'purchase_count', 'discount_rate', 'total_spent'
]].dropna().copy()

# --- Добавляем две отдельные бинарные колонки ---
df_clusters['education_высшее'] = (df_clusters['education'] == 'высшее').astype(int)
df_clusters['education_среднее'] = (df_clusters['education'] == 'среднее').astype(int)

# --- Удаляем оригинальный столбец ---
df_clusters.drop(columns=['education'], inplace=True)

# --- Масштабируем ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clusters)
```

```python
# --- Посмотрим форму масштабированных данных ---
print("Форма X_scaled:", X_scaled.shape)

# --- Сравним исходный датафрейм df_clusters ---
df_clusters.head()
```

```python
# --- Оценка качества кластеризации по силуэту ---
silhouette_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)
    print(f'k = {k}, silhouette score = {score:.4f}')

# --- График силуэтов ---
plt.figure(figsize=(8, 4))
plt.plot(K_range, silhouette_scores, marker='o')
plt.title('Silhouette Score для разных значений кластеров')
plt.xlabel('Число кластеров (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

```python
# --- Считаем инерцию для k от 1 до 10 ---
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# --- Строим график локтя ---
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Метод локтя — подбор количества кластеров')
plt.xlabel('Количество кластеров')
plt.ylabel('Инерция (сумма квадратов расстояний до центров)')
plt.grid(True)
plt.show()
```

# Модель склонности к покупке

Клиенты из страны 32, города 1188

```python
# --- Отбор обучающей выборки ---
df_ab = df_model[df_model['ab_group'].notna()].copy()
df_ab['target'] = (df_ab['ab_group'] == 'test').astype(int)

# --- Признаки (education будет закодирован отдельно) ---
base_features = ['age', 'gender_final', 'purchase_count', 'discount_rate', 'total_spent', 'cluster']
education_column = 'education'

# --- One-Hot Encoding education ---
df_ab_encoded = pd.get_dummies(df_ab[[education_column]], prefix='education')

# --- Сохраняем список финальных признаков ---
education_dummies = ['education_высшее', 'education_среднее']
for col in education_dummies:
    if col not in df_ab_encoded.columns:
        df_ab_encoded[col] = 0

# --- Собираем обучающую матрицу ---
X = pd.concat([df_ab[base_features], df_ab_encoded[education_dummies]], axis=1)
y = df_ab['target']

# --- Обучение модели ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = RandomForestClassifier(n_estimators=200, random_state=42)
model.fit(X_train_scaled, y_train)

# --- Оценка ---
y_pred = model.predict(X_test_scaled)
y_proba = model.predict_proba(X_test_scaled)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
```

## Вывод

Судя по метрикам:
- F1-score (для класса 1): 0.71 — модель неплохо справляется с выявлением тех, кто откликнулся на коммуникацию.
- Recall (для класса 1): 0.80 — высокая полнота, мы почти всех поймали.
- ROC AUC: 0.51 — на уровне случайного угадывания, сигнал слабый.

Модель ориентирована на максимальный захват откликнувшихся (recall), но теряет точность и плохо отделяет откликнувшихся от неоткликнувшихся (низкий AUC).

Это может быть следствием:
- небольшой обучающей выборки,
- слабой предиктивной силы признаков,
- высокой корреляции признака `cluster` с `target`, но недостаточной информативности остальных.
